```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
df.columns
```

```python
X = df[['Año', 'DOY', 'Hora', 'Minuto',"BZ, nT (GSM)",'ID for IMF spacecraft', 'BX, nT (GSE, GSM)', 'BY, nT (GSE)', 'Reconeccion_Magnetica']]
y = df['BZ, nT (GSE)']
```

```python
#X = X.values.reshape(-1, 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Crear un modelo de regresión lineal
modelo = LinearRegression()

# Entrenar el modelo
modelo.fit(X, y)
```

```python
y_pred = modelo.predict(X_test)

print("Predicciones: {}, Valores actuales: {}".format(y_pred[:5], y_test[:5]))
```

```python
modelo.score(X_train,y_train), modelo.score(X_test,y_test)
print(f"El R² para los datos de entrenamiento es: {modelo.score(X_train,y_train)} y para los de testeo: {modelo.score(X_test,y_test)}")
```

```python
# Imprimir los coeficientes de la regresión
print('Coeficiente de intercepción (b0):', modelo.intercept_)
print('Coeficientes de las características (b1, b2):', modelo.coef_)
```

```python
mse = mean_squared_error(y_test, y_pred)
print("Error cuadrático medio:", mse)
```
